```python
# Dados JSON importados do site Data.gov.sg (https://data.gov.sg/dataset/usage-of-mobile-devices-for-media-activities?resource_id=8dccb687-9265-415b-909a-86bcf578c22f)

import requests
import json
url = "https://data.gov.sg/api/action/datastore_search?resource_id=8dccb687-9265-415b-909a-86bcf578c22f&limit=10"
content = requests.get(url).content
unformatted_data = json.loads(content)
data = json.dumps(unformatted_data, indent=2)
print(data)
```
